```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
dfx = pd.read_csv("xdata.csv")
dfy = pd.read_csv("ydata.csv")
```

```python
x = dfx.to_numpy()
y = dfy.to_numpy()

x = x[:,1:]
y = y[:,1:]

y = y.reshape((-1,))

# print(x)
print(x.shape)
print(y.shape)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
query_x = np.array([2,3])
plt.scatter(x[:,0],x[:,1],c=y)
plt.scatter(query_x[0],query_x[1],c="red")
plt.show()
```

```python
def ecluidDist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

print(x[0],x[1])
print(ecluidDist(x[0],x[1]))
```

```python
def knn(x,y,qx,k=5):
    dist = []
    for i in range(x.shape[0]):
        d = ecluidDist(qx,x[i])
        dist.append((d,y[i]))
    dist = sorted(dist)
    dist = dist[:k]
    # print(dist)
    dist = np.array(dist)
    # print(dist)
    nv = np.unique(dist[:,1],return_counts=True)
    # print(nv)
    idx = nv[1].argmax()
    return nv[0][idx]

print(knn(x,y,query_x))
    
```
